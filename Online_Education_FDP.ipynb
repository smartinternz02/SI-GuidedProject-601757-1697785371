```python
corr = df[['Gender', 'Age(Years)', 'Time spent on social media (Hours)', 'Home Location']].corr()

sns.heatmap(corr, annot=True)

plt.title('Gender vs Age on Time spent on Social Media')

plt.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=df['Age(Years)'],
        y=df['Time spent on social media (Hours)'],
        name="Performance"
    ))
fig.add_trace(
    go.Bar(
        x=df['Age(Years)'],
        y=df['Study time (Hours)'],
        name="Subjects"
    ))
fig.show()
```
